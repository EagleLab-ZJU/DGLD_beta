```python
import pandas as pd
```

```python
res=pd.read_csv('./comga.csv')
```

```python
res[res['dataset']=='ACM'].sort_values('auc',ascending=False)
```

```python
def get_top(x):
    x=x.sort_values('auc',ascending=False)
    return x.iloc[0]
```

```python
drop_feas = ['logdir','device']
use_cols = [i for i in res.columns.values if i not in drop_feas]
```

```python
res.groupby('dataset',as_index=False).apply(get_top)[use_cols]
```
